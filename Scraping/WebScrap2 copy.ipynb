```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager
import os
import requests

# Setup Edge WebDriver
service = Service(EdgeChromiumDriverManager().install())
driver = webdriver.Edge(service=service)

for pag in range(2,11):
    # Open the target webpage
    driver.get(f"https://www.istockphoto.com/photos/frustrated-face?page={pag}")

    # Create a directory to save the images
    os.makedirs(f"downloaded_images_{pag}", exist_ok=True)

    # Locate all img elements with the specified class
    img_elements = driver.find_elements(By.CSS_SELECTOR, "img._BZ9oiIzKJoKWjkJDof_")

    for index, img in enumerate(img_elements):
        src = img.get_attribute("src")
        if src:
            try:
                response = requests.get(src, stream=True)
                if response.status_code == 200:
                    file_path = os.path.join(f"downloaded_images_{pag}", f"image_{index}.jpg")
                    with open(file_path, "wb") as file:
                        for chunk in response.iter_content(1024):
                            file.write(chunk)
                    print(f"Downloaded {file_path}")
                else:
                    print(f"Failed to download {src}: HTTP {response.status_code}")
            except Exception as e:
                print(f"Error downloading {src}: {e}")

driver.quit()
```

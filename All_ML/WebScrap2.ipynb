```python
import numpy as np
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
import os
```

```python
# Set the path to the directory containing the Microsoft Edge WebDriver executable
edge_driver_path = 'edgedriver_win64'
os.environ['PATH'] += ';' + edge_driver_path

# Create a new instance of the Edge driver
driver = webdriver.Edge()

# Open the website
driver.get("https://collegedunia.com/btech-colleges")

# Get the page source
html_source = driver.page_source

# Pass the HTML source to BeautifulSoup for parsing
soup = BeautifulSoup(html_source, 'html.parser')

# # Example: Print text of all <a> tags
# for a_tag in soup.find_all('a'):
#     print(a_tag.text.strip())

name=[]
fee_yr =[]
review_count=[]
rating=[]


for item in soup.find_all('tr',class_='jsx-3749532717'):
    name.append(item.find('h3').text.strip())

    fees_temp=item.find_all('span',class_='text-green')[0].text.strip()
    fees_temp=fees_temp.replace("â‚¹","")
    fees_temp=fees_temp.replace(" ","")
    fee_yr.append(fees_temp)

    rev_count = item.find('span',class_='lr-value').text.strip()
    rev_count = rev_count.replace("Based on ",'')
    rev_count = rev_count.replace(' UserReviews','')
    review_count.append(rev_count)

    rate=item.find('span',class_='lr-key').text.strip()
    rate=rate.replace(' / 5','')
    rating.append(rate)


# Close the browser
driver.quit()
```

```python
d= {
    'Name' : name,
    'Fee per year' : fee_yr,
    'Rating':rating,
    'Number of reviews':review_count
}
```

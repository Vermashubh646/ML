By using Albumentations

```python
import albumentations as A
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
transform = A.Compose([

    A.RandomCrop(width=256, height=256),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
    A.Affine(shear=5)

])
```

```python
image = cv2.imread('cat.jpeg')
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# convert the image from BGR to RGB
```

```python
transformed_1 = transform(image=image)["image"]
transformed_2 = transform(image=image)['image']
transformed_3 = transform(image=image)['image']
```

```python
plt.imshow(transformed_1[:,:,:])
plt.show()
```

```python
plt.imshow(transformed_2)
plt.show()
```

```python
plt.imshow(transformed_3)
plt.show()
```
